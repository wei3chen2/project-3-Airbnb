```python
# load necessary libraries
import pandas as pd
import pprint
import os
import scipy.stats as st
from scipy.stats import linregress

```

```python
# import csvs
paris_csv = "data/Paris.csv"
bangkok_csv = "data/Bangkok.csv"
newyork_csv = "data/NewYork.csv"
london_csv = "data/London.csv"
hongkong_csv = "data/HongKong.csv"
singapore_csv = "data/Singapore.csv"
instanbul_csv = "data/Instanbul.csv"
```

```python
# read csv into pandas df
paris_df = pd.read_csv(paris_csv)
bangkok_df = pd.read_csv(bangkok_csv)
newyork_df = pd.read_csv(newyork_csv,low_memory=False)
london_df = pd.read_csv(london_csv)
hongkong_df = pd.read_csv(hongkong_csv)
singapore_df = pd.read_csv(singapore_csv)
instanbul_df = pd.read_csv(instanbul_csv)
```

```python
# add "city" column to each df 
paris_df["city"] = "paris"
bangkok_df["city"] = "bangkok"
newyork_df["city"] = "newyork"
london_df["city"] = "london"
hongkong_df["city"] = "hongkong"
singapore_df["city"] = "singapore"
instanbul_df["city"] = "instanbul"
```

```python
# convert price to USD
 
# conversion rate (as of 2023-03-30)

    # London (pounds) 1 pound = 1.24 usd
    # Paris (euro) 1 euro = 1.09 usd
    # Instanbul (lira) 1 lira = 0.052 usd
    # Bangkok (thai baht) 1 thai baht = 0.029 usd
    # Hong Kong (hong kong dollar) 1 hong kong dollar = 0.13 usd
    # Singapore (singapore dollar) 1 singapore dollar = 0.75 usd
```

```python
# add new york price to new column
newyork_df['price'] = newyork_df['price'].str.replace(',', '').str.replace('$', '')
```

```python
newyork_df['USD_price'] = newyork_df['price']

```

```python
# convert london
london_df['price'] = london_df['price'].str.replace(',', '').str.replace('$', '')

london_df['price'] = pd.to_numeric(london_df['price'])
london_df['USD_price'] = london_df['price']*1.24
```

```python
# convert instanbul
instanbul_df['price'] = instanbul_df['price'].str.replace(',', '').str.replace('$', '')
```

```python
# convert singapore
singapore_df['price'] = singapore_df['price'].str.replace(',', '').str.replace('$', '')
singapore_df['price'] = pd.to_numeric(singapore_df['price'])
singapore_df['USD_price'] = singapore_df['price']*0.75
```

```python
# filter by room type
clean_df = cleaned_df.loc[cleaned_df["room_type"].isin(["Private room", "Entire home/apt"])]
clean_df.head()
```

```python
# rename columns 
clean_df.columns = ['Id', 'Listing_Url', 'Name', 'Picture_Url', 'Host_Name', 'Host_Since', 'Host_Response_Rate',
       'Superhost', 'Neighbourhood', 'Latitude', 'Longitude', 'Room_Type', 'Accommodates', 'Bathrooms', 'Bedrooms', 
        'Beds', 'Amenities', 'Review_Scores_Rating', 'Review_Scores_Cleanliness', 'Review_Scores_Checkin',
        'Review_Scores_Communication', 'Review_Scores_Location', 'Review_Scores_Value', 'City', 'Price(USD)']
clean_df.head()
```

```python
# export cleaned_df (encoding because of french characters)
clean_df.to_csv("clean.csv", index=False, header=True, encoding='utf-8-sig')
```

```python
#generate the dataframe based on the room type for each city 

room_type_city=clean_df['Room_Type'].groupby(clean_df['City']).value_counts()
room_type_city_df = pd.DataFrame(room_type_city)
room_type_city_df
```

```python
# export cleaned_df (encoding because of french characters)
room_type_city.to_csv("room_type_city.csv", index=True, header=True,sep=',')
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the price for each city

price_mean = clean_df['Price(USD)'].groupby(clean_df['City']).mean()
price_median = clean_df['Price(USD)'].groupby(clean_df['City']).median()
price_var = clean_df['Price(USD)'].groupby(clean_df['City']).var()
price_std = clean_df['Price(USD)'].groupby(clean_df['City']).std()
price_sem = clean_df['Price(USD)'].groupby(clean_df['City']).sem()

# Assemble the resulting series into a single summary DataFrame.

summary_stat_df = pd.DataFrame({"Mean price":price_mean, 
                            "Median price":price_median, 
                           "Price Variance":price_var, 
                           "Price Std. Dev.":price_std, 
                           "Price Std. Err.":price_sem})
summary_stat_df
```

```python
# export cleaned_df (encoding because of french characters)
summary_stat_df.to_csv("summary_stat_df.csv", index=True, header=True,columns=None,encoding='utf-8-sig')
```

```python
# Using the aggregation method, produce the same summary statistics in a single line.

summary_aggr_df=clean_df.groupby("City")[["Price(USD)"]].agg(["mean", "median", "var", "std", "sem"])
summary_aggr_df
```
